```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm  # scipy が無い場合は下記の手動計算でも可

mu = 0.0      # 平均
sigma = 1.0   # 標準偏差

x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, loc=mu, scale=sigma)

plt.plot(x, y, label=f"μ={mu}, σ={sigma}")
plt.fill_between(x, y, alpha=0.2)
plt.title("正規分布")
plt.xlabel("x")
plt.ylabel("確率密度")
plt.legend()
plt.grid(True)
plt.show()
```
